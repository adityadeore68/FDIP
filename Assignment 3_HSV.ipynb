```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
```

```python
url = "https://images.unsplash.com/photo-1506744038136-46273834b3fb"
resp = urllib.request.urlopen(url)
image_array = np.asarray(bytearray(resp.read()), dtype=np.uint8)
img_bgr = cv2.imdecode(image_array, cv2.IMREAD_COLOR)
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(img_rgb)
plt.title("Original (from URL)")
plt.axis('off')
plt.show()
```

```python
hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,5))
ax[0].imshow(h, cmap="gray"); ax[0].set_title("Hue"); ax[0].axis("off")
ax[1].imshow(s, cmap="gray"); ax[1].set_title("Saturation"); ax[1].axis("off")
ax[2].imshow(v, cmap="gray"); ax[2].set_title("Value"); ax[2].axis("off")
plt.show()
```

```python
ycrcb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2YCrCb)
Y, Cr, Cb = cv2.split(ycrcb)
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,5))
ax[0].imshow(Y, cmap="gray"); ax[0].set_title("Y (Luma)"); ax[0].axis("off")
ax[1].imshow(Cb, cmap="gray"); ax[1].set_title("Cb"); ax[1].axis("off")
ax[2].imshow(Cr, cmap="gray"); ax[2].set_title("Cr"); ax[2].axis("off")
plt.show()
```
